```python
import sys
ar=[-1,2,1,4]
target = 1
closesum = sys.maxsize
ar.sort()

for i in range(len(ar)-2):
  l = i+1
  r = len(ar) - 1
  while l<r:
    temp_sum = ar[i]+ar[l]+ar[r]

    if (abs(target - temp_sum) < abs(target - closesum)) :
      closesum = temp_sum;
    if temp_sum > target:
      r=r-1
    if temp_sum < target:
      l = l+1
print(closesum)

```

```python
def nextPermutation(arr):
  n = len(arr)
  i = 0
  j = 0
  for i in range(n-2, -2, -1):
    if i!=-1:
      if (arr[i] < arr[i + 1]):
        break

  # Check if pivot is not found
  if (i < 0):
    arr.reverse()

  # if pivot is found
  else:
    # Find for the number > pivot 
    for j in range(n-1, i, -1):
      if (arr[j] > arr[i]):
        break

    # swap elements
    arr[i], arr[j] = arr[j], arr[i]     
    # range of suffix

    strt, end = i+1, len(arr)

    # reverse the string
    arr[strt:end] = arr[strt:end][::-1]
  return arr

nextPermutation([1,2,3])
```

```python
def search_insert(nums, target):

  # Last and First indexes
  start = 0
  end = len(nums) - 1

  # Traverse an array
  while (start <= end):
      
      mid = (start + end)//2
        
      # if target value found.
      if nums[mid] == target:
          return mid
      
      # If target value is greater then mid elements's value
      elif target > nums[mid]:
          start = mid + 1
          
      # otherwise target value is less, 
      else:
          end = mid -1
  # Return the insertion position
  return end + 1

search_insert([1,3,5,6],5)
```
