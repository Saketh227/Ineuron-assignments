```python
def fun(matrix,target):
  start_row = 0
  end_row = len(matrix) -1
  end_col = len(matrix[0]) - 1

  while start_row <= end_row:
    mid = start_row +((end_row-start_row)//2)
    if matrix[mid][0] > target:
      end_row = mid -1
    elif matrix[mid][0] < target:
      if target <= matrix[mid][end_col]:
        start_row = mid
        if target in matrix[start_row]:
          return True
        else:
          return False
      else:
        start_row = mid+1
    elif matrix[mid][0] == target:
      return True

 

fun([[1,3,5,7],[10,11,16,20],[23,30,34,60],[70,71,72,73]],3)
```

```python
def fun(nums):
  count = 0
        
  map = { 0: -1}
  
  max_length = 0
  
  for i, number in enumerate(nums):
    if number:
      count += 1
    else:
      count -= 1

    if count in map:
      max_length = max( max_length, ( i - map[count] ) )   
    else:
      map[ count ] = i
          
  return max_length

  
fun([1, 0, 0, 1, 0, 1, 1])
```

```python
def fun(nums1,nums2):
  nums1.sort()
  nums2.sort()
  n, res = len(nums1), 0
  for i in range(n):
      res += nums1[i] * nums2[n - i - 1]
  return res

fun([5,3,4,2],[4,2,2,5])
```

```python
import numpy as np
def get_spiral(n):
  array = np.ones((n,n))
  count=0
  up=0
  left=0
  right = n-1
  down = n-1

  while up<=down & left<=right:
    # Traverse from left to right
    for i in range(left,right+1):
      count += 1
      array[up][i] = count
    
    # Update the up pointer
    up = up+1

    # Traverse from up to down
    for i in range(up,down+1):
      count += 1
      array[i][right] = count
    
    # Update right pointer
    right = right-1
    
    # Traverse from right to left
    if up <= down:
      for i in range(right,left-1,-1):
        count += 1
        array[down][i] = count

      # Update down pointer
      down = down-1

    # Traverse from down to up
    if left <= right:
      for i in range(down,up-1,-1):
        count += 1
        array[i][left] = count 

      # Update down pointer
      left = left+1
  
  
  return array

get_spiral(3)
```

```python

def multiply(mat1,mat2):
  m = len(mat1)
  n = len(mat2)
  l = len(mat2[0])
  ans = [[0] * l for _ in range(m)]

  for i in range(m):
    for j in range(l):
      for k in range(n):
        ans[i][j] += mat1[i][k] * mat2[k][j]

  return ans

mat1 = [[1,0,0],[-1,0,3]]
mat2 = [[7,0,0],[0,0,0],[0,0,1]]

multiply(mat1,mat2)

```
