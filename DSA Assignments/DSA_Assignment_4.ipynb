```python
def fun(arr1,arr2,arr3):
  '''
  The three arrays are already sorted in increasing order
  so we shall traverse in one array and 
  look if the element is present in both the other arrays
  if the above condition is satisfied add the element to the result.
  Because they are already sorted the result will be in increasing order only
  '''
  res=[]
  for i in arr1:
    if (i in arr2) & (i in arr3):
      res.append(i)
  return res

fun([1,2,3,4,5],[1,2,5,7,9],[1,3,4,5,8])
```

```python
def find_common(arr1, arr2, arr3):
    i, j, k = 0, 0, 0
    n1,n2,n3 = (len(arr1),len(arr2),(len(arr3)))
    res=[]
 
    # Iterate through three arrays while all arrays have elements
    while (i < n1 and j < n2 and k < n3):
 
        # If x = y and y = z, print any of them and move ahead
        # in all arrays
        if (arr1[i] == arr2[j] and arr2[j] == arr3[k]):
            res.append(arr1[i])
            i += 1
            j += 1
            k += 1
 
        # x < y
        elif arr1[i] < arr2[j]:
            i += 1
 
        # y < z
        elif arr2[j] < arr3[k]:
            j += 1
 
        # We reach here when x > y and z < y, i.e., z is smallest
        else:
            k += 1
    return res
find_common([1,2,3,4,5],[1,2,5,7,9],[1,3,4,5,8])
```

```python
def get_distinct(arr1,arr2):
  res=[[i for i in set(arr1) if i not in arr2],[i for i in set(arr2) if i not in arr1]]
  return res
get_distinct([1,2,3],[2,4,6])
```

```python
import numpy as np
def matrix_transpose(array):
  rows,cols = np.array(array).shape
  res=[]
  for i in range(cols):
    temp=[]
    for j in range(rows):
      temp.append(array[j][i])
    res.append(temp)
  return res
matrix_transpose([[1,2,3],[4,5,6],[7,8,9]])
```

```python
def get_max_sum(array):
  array.sort()
  return sum([array[i] for i in range(len(array)) if i%2 ==0 ])
get_max_sum([1,2,3,4])
```

```python
def n_rows(n):
  row=0
  i=1
  while i<=n:
    n=n-i
    row += 1
    i += 1
  return row

n_rows(7)
```

```python
import math 
'''
using sum of A.P formula
s = (n/2)[2a + (n-1)d]
2s = n[2*1 + (n-1)*1]
2s = n[2+n-1]
2s = n[n+1]
solve above eq
where s is target n is no.of rows
'''
def n_rows(n):
  return int((-1 + math.sqrt(1+8*n))/2)
n_rows(7)
```

```python
def sorted_square(array):
  res=[]
  start=0
  end = len(array)-1
  while start <=end:
    if abs(array[start]) < abs(array[end]):
      res.append(array[end]**2)
      end -= 1
    else:
      res.append(array[start]**2)
      start += 1
  return res[::-1]

sorted_square([-4,-1,0,3,10])
```

```python
def max_int(m,n,ops):
  min_a = m
  min_b = n
  for op in ops:
    if op[0] < min_a :
      min_a = op[0]
    if op[1] < min_b:
      min_b = op[1]
  # print(min_a,min_b)
  return (min_a)*(min_b)

max_int(3,3,[[2,2],[3,3]])
```

```python
def transform(array,n):
  res=[]
  for i in range(n):
    res.extend([array[i],array[n+i]])
  return res

transform([2,5,1,3,4,7],3)
```
