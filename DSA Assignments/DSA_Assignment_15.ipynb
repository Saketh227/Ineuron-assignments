```python
def fun(arr,n):
  ans=[]
  for i in range(n):
    next = -1
    for j in range(i+1,n):
      if arr[i] < arr[j]:
        next = arr[j]
        break
    ans.append(next)
  return ans

arr=[1,3,2,4]
fun(arr,len(arr))
```

```python
''' explanation :
the idea here is to compare each and every element in array with top of stack
when stack is empty add current element to stack and move on to next element in array

    a = [12,2,10,13]
    stack =[]
    ans =[-1 for _ in range(len(a))]

    step 1 :
      i=0
      stack is empty
      s=[12]
    step 2:
      i=1
      12 > a[1]
      s = [12,2]
    step 3:
      i=2
      2 < a[2]

'''



def nextGreaterElement(a):
    # initializing an empty stack
    s = []
    ans =[-1 for _ in range(len(a))]

    for i in range(len(a)):
        # If the stack is empty, push the current element.
        if len(s) == 0:
            s.append(a[i])
            continue

        while s and s[-1] < a[i]:
            ans[a.index(s[-1])] = a[i]
            s.pop()

        s.append(a[i])
    return ans



a = [12,2,10,13]
print("The list of the next greater element is :")
nextGreaterElement(a)
```

```python
# implement Stack using Queues

from collections import deque
class MyStack:

    def __init__(self):
        self.qa = deque()
        self.qb = deque()

    def push(self, x: int) -> None:
        self.qa.append(x)

    def pop(self) -> int:
        self.flush()
        val = self.qa.popleft()
        self.swap()

        return val

    def top(self) -> int:
        self.flush()
        return self.qa[0]

    def empty(self) -> bool:
        return len(self.qa) < 1 and len(self.qb) < 1

    def flush(self):
        while len(self.qa) > 1:
            self.qb.append(self.qa.popleft())

    def swap(self):
        self.qa, self.qb = self.qb, self.qa

stack = MyStack()
stack.push(2)
stack.push(3)
print(stack.qa)
print(stack.pop())
print(stack.qa)
stack.push(4)
print(stack.qa)
print(stack.pop())
print(stack.qa)
```

```python
def reverse(s):
  ans = []
  if len(s) == 0:
    return None
  if len(s) == 1:
    return s
  ans.append(s.pop())
  ans = ans + reverse(s)

  return ans

reverse([1,2,3])
```

```python
def findWater(arr, n):
	left = [0]*n
	right = [0]*n
	water = 0

	left[0] = arr[0]
	for i in range(1, n):
		left[i] = max(left[i-1], arr[i])

	right[n-1] = arr[n-1]
	for i in range(n-2, -1, -1):
		right[i] = max(right[i + 1], arr[i])

	for i in range(0, n):
		water += min(left[i], right[i]) - arr[i]

	return water

arr = [4,2,0,3,2,5]
n = len(arr)
print(findWater(arr, n))

```
