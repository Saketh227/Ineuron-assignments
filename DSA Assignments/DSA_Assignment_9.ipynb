```python
# Power of 2
def check(n):
  # base condition 1 = 2^0
  if n == 1:
    return True
  elif n%2 != 0 or n<=0:
    return False


  else:
    return check(n/2)

check(16)
```

```python
def get_sum(n):
  if n==1 or n==0:
    return n
  else:
    return n+get_sum(n-1)

get_sum(5)
```

```python
def factorial(n):
  if n==0 | n==1:
    return 1
  else:
    return n*factorial(n-1)

factorial(4)
```

```python
def find_max(array):
  if len(array) == 1:
    return array[0]
  else:
    return max(array[0],find_max(array[1:]))

find_max([1,4,3,-5,-4,8,6])
```

```python
def get(a,d,N):
  if N==1:
    return a
  else:
    return get(a+d,d,N-1)

get(5,2,10)
```

```python
def permutation(s):
  if len(s) == 1:
    return [s]
  else:
    res = []
    for i, c in enumerate(s):
      for perm in permutation(s[:i] + s[i+1:]):
        res.append(c + perm)
    return res

permutation('ABC')
```

```python
def prod(array):
  # base case
  if len(array) ==1:
    return array[0]

  elif len(array)==0:
    return None
  else:
    return (array[0] * prod(array[1:]))

prod([1,6,3])
```
