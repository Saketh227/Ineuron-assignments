```python
def merge_intervals(intervals:list)->list:
    intervals.sort(key=lambda x:x[0])

    curr = [intervals[0]]
    for i in range(1,len(intervals)):
        if intervals[i][0] <= curr[-1][1] and curr[-1][1] < intervals[i][1]:
            curr[-1] = [curr[-1][0] ,intervals[i][1] ]
        else :
            curr.append(intervals[i])
    print(curr)

merge_intervals([[1,3],[8,10],[2,6],[15,18]])
merge_intervals(([[1,4],[4,5]]))
```

```python
def sort_colors(l:list)->list:
    i=0
    while i < len(l):
        if l[i] == 0:
            del l[i]
            l = [0] +l

        elif l[i]==2:
            del l[i]
            l.append(2)

        i=i+1
    print(l)

sort_colors([2,0,2,1,1,0])
sort_colors([2,0,1])
```

```python
class version:
  def __init__(self,n,bad):
    self.n = n
    self.bad = bad
  def isBadVersion(self,val):
    if val >= self.bad:
      return True
    else:
      return False
  def get_bad_version(self):
    return self.bad

obj = version(5,4)
print(obj.isBadVersion(3),obj.isBadVersion(5),obj.isBadVersion(4),sep='\n')
```

```python
def getMaxGap(nums):
  nums.sort()
  if len(nums)<2:
    return 0
  maximum = 0
  for i in range(1,len(nums)):
    temp_max = nums[i]-nums[i-1]

    maximum = max(maximum,temp_max)

  print(maximum)

getMaxGap([3,6,9,1])
getMaxGap([0])
```

```python
def hasDuplicates(nums):
  d={}
  for i in nums:
    if i in d:
      return True
    else:
      d[i]=1
  return False

print(hasDuplicates([1,2,3,4]))
print(hasDuplicates([1,2,3,1]))
print(hasDuplicates([1,1,1,3,3,4,3,2,4,2]))
```

```python
def getArrows(l):
  '''
  just find intersections
  similar to finding merging intervals problem
  '''
  l.sort(key=lambda x:x[0])

  curr = [l[0]]
  for i in range(1,len(l)):
      if l[i][0] <= curr[-1][1] and curr[-1][1] < l[i][1]:
          curr[-1] = [ l[i][0],curr[-1][1] ]
      else :
          curr.append(l[i])
  print(len(curr))

getArrows([[10,16],[2,8],[1,6],[7,12]])
getArrows([[1,2],[3,4],[5,6],[7,8]])
getArrows([[1,2],[2,3],[3,4],[4,5]])
```

```python
def lengthOfLIS(nums):
  n = len(nums)
  ans = [1] * n
  for i in range(n):
    for j in range(i):
      if nums[i] > nums[j] and ans[i] < ans[j] + 1:
        ans[i] = ans[j] + 1
  return max(ans)

lengthOfLIS([10,9,2,5,3,7,101,18]),lengthOfLIS([0,1,0,3,2,3]),lengthOfLIS([7,7,7,7,7,7,7])
```

```python

def get(arr, n) :
	small = []
	min1 = arr[0]
	for i in range(n) :
		if (min1 >= arr[i]) :
			min1 = arr[i]
			small.append(-1)

		else :
			small.append(min1)

	s = []
	for i in range(n - 1, 0, -1) :
		while (len(s) != 0 and s[-1] <= small[i]) :
			s.pop()
		if (len(s) != 0 and small[i] != -1 and s[-1] < arr[i]) :
			return True

		s.append(arr[i])

	return False


arr = [1,2,3,4 ]
N = len(arr)
print(get(arr, N))
```
