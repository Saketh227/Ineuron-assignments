```python

def is_isomorphic(s,t):
  d={}
  res = True

  if len(s)!= len(t):
    res = False

  for i in range(len(s)):
    #create key, value pair if key is not present
    # else check if the value is same as current value
    if s[i] not in d.keys():
      d[s[i]] = t[i]
    else:
      if d[s[i]] == t[i]:
        continue
      else:
        res = False
        break
  return res

s="egg"
t='add'
is_isomorphic(s,t)
```

```python
# strobogrammatic
d={'0':'0','1':'1','6':'9','8':'8','9':'6'}
def is_strobogramatic(arr):
  start = 0
  end = len(arr)-1
  while start < end:
    if arr[start] in d.keys() and arr[end] in d.keys():
      if arr[start] == d[arr[end]]:
        start += 1
        end -= 1

      else:
        return False
    else:
      return False
  return True

is_strobogramatic('123')
```

```python
def add(nums1,nums2):
  l1 = len(nums1)
  l2 = len(nums2)
  res=''
  nums1 = nums1[::-1]
  nums2 = nums2[::-1]

  if l1 < l2:
    t = nums1
    nums1 = nums2
    nums2 =t
  carry = 0
  l1 = len(nums1)
  l2 = len(nums2)
  # nums1 has high length l1
  for i in range(l2):
    sum = (ord(nums1[i]) - 48) + (ord(nums2[i]) - 48) + carry
    res += chr(sum % 10 + 48)
    carry = int(sum / 10)

  for i in range(l2,l1):
    sum = (ord(nums1[i]) - 48) + carry
    res += chr(sum % 10 + 48)
    carry = int(sum / 10)

  # Add remaining carry
  if carry:
      res += chr(carry + 48)

  res = res[::-1]
  return res

add('99','999')
```

```python
def get(s):
  res=[]
  for i in s:
    if i!= '#':
      res.append(i)
    else:
      if len(res)!=0:
        res.pop(-1)
  return "".join(res)

def cmp(s1,s2):
  if get(s1) == get(s2):
    return True
  else:
    return False

cmp("ab#c","ad#c")
```
