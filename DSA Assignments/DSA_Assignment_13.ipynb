```python
class node:
  def __init__(self,val):
    self.val =val
    self.next = None
    self.curr = None

class linkedlist:
  def __init__(self):
    self.head = None
    self.count = 0
  def push(self,val):
    if self.count == 0:
      self.head = node(val)
      self.curr = self.head
    else:
      temp = self.curr
      self.curr = node(val)
      temp.next = self.curr
    self.count += 1

  def get_ll(self):
    temp = self.head
    if self.head == None:
      print("the list is empty")
    while temp.next != None:
      print(temp.val,end="->")
      temp = temp.next
    print(temp.val,end="->null\n")
```

```python
def combine(l1:linkedlist,l2:linkedlist)-> linkedlist:
  temp1 = l1.head
  temp2 = l2.head
  res = linkedlist()
  while temp1 != None and temp2 != None:
    res.push(max(temp1.val,temp2.val))

    temp1 = temp1.next
    temp2 = temp2.next

  return res

l1 = linkedlist()
l1.push(5)
l1.push(2)
l1.push(3)
l1.push(8)

l2 = linkedlist()
l2.push(1)
l2.push(7)
l2.push(4)
l2.push(5)

combine(l1,l2).get_ll()
```

```python
def reverse(curr, k):
  prev = None
  while curr and k > 0:
    next = curr.next
    curr.next = prev
    prev = curr
    curr = next

    k = k - 1
  return prev, curr


def skipKNodes(curr, k):
  while curr and k > 0:
    curr = curr.next
    k = k - 1
  return curr

def reverseAlternatingKNodes(head, k):
  if head is None:
    return None

  originalHead = head
  head, curr = reverse(head, k)
  originalHead.next = curr
  curr = skipKNodes(curr, k - 1)
  if curr:
    curr.next = reverseAlternatingKNodes(curr.next, k)
  return head



ll = linkedlist()
ll.push(1)
ll.push(2)
ll.push(2)
ll.push(4)
ll.push(5)
ll.push(6)
ll.push(7)
ll.push(8)
head = reverseAlternatingKNodes(ll.head,3)

while head != None:
  print(head.val,end='->')
  head = head.next
print('null')
```

```python
def mergeLists(headA, headB):
	dummyNode = node(0)
	tail = dummyNode
	while True:
		if headA is None:
			tail.next = headB
			break
		if headB is None:
			tail.next = headA
			break

		if headA.val <= headB.val:
			tail.next = headA
			headA = headA.next
		else:
			tail.next = headB
			headB = headB.next

		tail = tail.next

	return dummyNode.next


listA = linkedlist()
listB = linkedlist()

listA.push(5)
listA.push(10)
listA.push(15)

listB.push(2)
listB.push(3)
listB.push(20)

listA.head = mergeLists(listA.head, listB.head)

listA.get_ll()
```

```python
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None
		self.prev = None


class DoublyLinkedList:
	def __init__(self):
		self.head = None

	def reverse(self):
		temp = None
		current = self.head

		while current is not None:
			temp = current.prev
			current.prev = current.next
			current.next = temp
			current = current.prev

		if temp is not None:
			self.head = temp.prev

	def push(self, new_data):
		new_node = Node(new_data)
		new_node.next = self.head

		if self.head is not None:
			self.head.prev = new_node
		self.head = new_node

	def printList(self, node):
		while(node is not None):
			print(node.data, end=' ')
			node = node.next


dll = DoublyLinkedList()
dll.push(2)
dll.push(4)
dll.push(8)
dll.push(10)

print("\nOriginal Linked List")
dll.printList(dll.head)

# Function call
dll.reverse()

print("\nReversed Linked List")
dll.printList(dll.head)


```

```python
class Node:
	def __init__(self, data):
		self.data = data
		self.next = None
		self.prev = None


class DoublyLinkedList:
	def __init__(self):
		self.head = None

	def reverse(self):
		temp = None
		current = self.head

		while current is not None:
			temp = current.prev
			current.prev = current.next
			current.next = temp
			current = current.prev

		if temp is not None:
			self.head = temp.prev

	def push(self, new_data):
		new_node = Node(new_data)
		new_node.next = self.head

		if self.head is not None:
			self.head.prev = new_node
		self.head = new_node

	def printList(self, node):
		while(node is not None):
			print(node.data, end=' ')
			node = node.next


def deleteNode(head_ref, del_):
	if (head_ref == None or del_ == None):
		return

	if (head_ref == del_):
		head_ref = del_.next

	if (del_.next != None):
		del_.next.prev = del_.prev

	if (del_.prev != None):
		del_.prev.next = del_.next

	return head_ref

def deleteNodeAtGivenPos(head_ref,n):
	if (head_ref == None or n <= 0):
		return

	current = head_ref
	i = 1
	while ( current != None and i < n ):
		current = current.next
		i = i + 1

	if (current == None):
		return
	deleteNode(head_ref, current)

	return head_ref

dll = DoublyLinkedList()
dll.push(2)
dll.push(4)
dll.push(8)
dll.push(10)


dll.printList(dll.head)

# Function call
head = deleteNodeAtGivenPos(dll.head,2)

dll.printList(head)
```
