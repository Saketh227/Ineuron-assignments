```python
def value(r):
	if (r == 'I'):
		return 1
	if (r == 'V'):
		return 5
	if (r == 'X'):
		return 10
	if (r == 'L'):
		return 50
	if (r == 'C'):
		return 100
	if (r == 'D'):
		return 500
	if (r == 'M'):
		return 1000
	return -1

def romanToDecimal(str):
	res = 0
	i = 0

	while (i < len(str)):
		s1 = value(str[i])
		if (i + 1 < len(str)):
			s2 = value(str[i + 1])
			if (s1 >= s2):
				res = res + s1
				i = i + 1
			else:
				res = res + s2 - s1
				i = i + 2
		else:
			res = res + s1
			i = i + 1

	return res

print(romanToDecimal("III"))
```

```python
def areDistinct(str, i, j):
	visited = [0] * (26)

	for k in range(i, j + 1):
		if (visited[ord(str[k]) -
				ord('a')] == True):
			return False

		visited[ord(str[k]) -
				ord('a')] = True

	return True

def longestUniqueSubsttr(str):
	n = len(str)
	res = 0
	for i in range(n):
		for j in range(i, n):
			if (areDistinct(str, i, j)):
				res = max(res, j - i + 1)
	return res

my_str = "abcabcbb"
print(longestUniqueSubsttr(my_str))
```

```python
def group_anagrams(words):
  anagram_dict = {}
  for word in words:
    # print(sorted(word))
    sorted_word = ''.join(sorted(word))
    if sorted_word in anagram_dict:
      anagram_dict[sorted_word].append(word)
    else:
      anagram_dict[sorted_word] = [word]
  return list(anagram_dict.values())


words = ["eat","tea","tan","ate","nat","bat"]
print(group_anagrams(words),group_anagrams(["a"]),sep="\n")
```

```python
def maxDivide(a, b):
	while a % b == 0:
		a = a / b
	return a

def isUgly(no):
	no = maxDivide(no, 2)
	no = maxDivide(no, 3)
	no = maxDivide(no, 5)
	return 1 if no == 1 else 0

def getNthUglyNo(n):
	i = 1
	count = 1
	while n > count:
		i += 1
		if isUgly(i):
			count += 1
	return i

getNthUglyNo(10)
```

```python
def findCrossOver(arr, low, high, x) :
	if (arr[high] <= x) :
		return high

	if (arr[low] > x) :
		return low

	mid = (low + high) // 2

	if (arr[mid] <= x and arr[mid + 1] > x) :
		return mid
	if(arr[mid] < x) :
		return findCrossOver(arr, mid + 1, high, x)

	return findCrossOver(arr, low, mid - 1, x)

def printKclosest(arr, x, k, n) :
  res=[]
  l = findCrossOver(arr, 0, n - 1, x)
  print(l)
  res.append(arr[l])
  r = l + 1
  count = 1
  l -= 1
  while (l >= 0 and r < n and count+2 <= k) :
    res.append(arr[l])
    res.append(arr[r])
    l -= 1
    r += 1
    count += 2
  while count != k :
    if l>=0 and(x-arr[l] < arr[r]-x or x-arr[l] == arr[r]-x):
      res.append(arr[l])
      l=l-1
    else :
      res.append(arr[r])
      r=r+1
    count +=1

  return res


arr =[1,2,3,4,5]
n = len(arr)
x = 3
k = 4

printKclosest(arr, x, k, n), printKclosest(arr, -1, k, n)
```
