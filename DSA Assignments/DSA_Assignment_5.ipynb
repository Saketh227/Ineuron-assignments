```python
def n_rows(n):
  row=0
  i=1
  while i<=n:
    n=n-i
    row += 1
    i += 1
  return row

n_rows(5)
```

```python
import math 
'''
using sum of A.P formula
s = (n/2)[2a + (n-1)d]
2s = n[2*1 + (n-1)*1]
2s = n[2+n-1]
2s = n[n+1]
solve above eq
where s is target n is no.of rows
'''
def n_rows(n):
  return int((-1 + math.sqrt(1+8*n))/2)
n_rows(5)
```

```python
def sorted_square(array):
  res=[]
  start=0
  end = len(array)-1
  while start <=end:
    if abs(array[start]) < abs(array[end]):
      res.append(array[end]**2)
      end -= 1
    else:
      res.append(array[start]**2)
      start += 1
  return res[::-1]

sorted_square([-4,-1,0,3,10])
```

```python
# Brute Force

def get_distance(arr1,arr2,d):
  count = 0
  for i in arr1:
    ''' 
        Time complexity : 
        Space Complexity :
        ******************************************
        ******************************************
        checking if the absolute difference b/w
        one element from arr1 and 
        each and every element form arr2
        is > d

        if > d
            return True
        else 
            False

        if all values turn out to be true 
        then increase the count
    '''
    if all([abs(i-j) > d for j in arr2]):
      count += 1
  return count

get_distance([4,5,8],[10,9,1,8],2)
```

```python
def get_distance(arr1,arr2,d):
  count = 0
  arr2.sort()
  for i in arr1:
    if min(abs(i-arr2[0]),abs(i-arr2[-1])) <= d:
      continue
    else:
      count += 1
  return count
get_distance([4,5,8],[10,9,1,8],2)
```

```python
def get_duplicate_values(arr):
  res=[]
  arr.sort()
  for i in range(len(arr)-1):
    if arr[i] == arr[i+1]:
      res.append(arr[i])
  return res
get_duplicate_values([4,3,2,7,8,2,3,1])
```
