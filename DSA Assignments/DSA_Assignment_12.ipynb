# <b>List Vs Linked list Vs array</b>

* Array has **`contiguous memory`**
* Array has **`fixed size`**

<hr>

## List:
* To avoid fixed size of array we use lists

* List has **`contiguous memory`** and has **`no fixed size`**

* In the background list is actually an array. so we can use list to `access elements` or `add elements at the end`

### When to use lists?

* Use lists when you just need to
      1.Acess elements
      2.Insert elemets at end

* It is **`slower`** when we try to **`insert at the beginning or in the middle`**. Because all **`other elements need to shifted`**.

<hr>

## Linked List

* To avoid drawbacks in list we use linked list.

* It has **`no contiguous memory and no fixed size`**.

## When to use linked list?
* use this when we want to insert elements.

# Implementation of linked list in python

```python
class node:
  def __init__(self,val):
    self.val = val
    self.next = None


class Linkedlist:

  def __init__(self):
    self.head = 'null'
    self.count = 0
    self.curr = None

  def append(self,val):
    # When linked list is empty
    if self.count == 0:
      self.head = node(val)
      self.curr = self.head
      self.count +=1
    # When adding 2 or above elements
    else:
      temp = self.curr
      self.curr = node(val)
      temp.next = self.curr
      self.count += 1

  def get_count(self):
    print(f"The no. elements in the given list are {self.count}")
    return self.count

  def get_ll(self):
    temp = self.head
    if self.head == None:
      print("the list is empty")
    while temp.next != None:
      print(temp.val,end="->")
      temp = temp.next
    print(temp.val,end="->null\n")



  def remove(self, target:int):
    print(f"The linkedd list before removing {target}:",end =" ")
    self.get_ll()
    temp = self.head

    if self.count ==1 and  temp.val == target:
      self.head = None
      return
    else:
      return

    while temp.next != None:
      # if the value is at start
      if temp == self.head and temp.val == target:
        self.head = temp.next
        temp.next = None
        self.count -=1
        return

      # if value is at last
      elif temp.next == self.curr and temp.next.val == target:
        # print("removing last element")
        temp.next = None
        self.curr = temp
        self.count -=1
        return

      # if value is in middle
      elif temp.next.val == target:
        dummy = temp.next
        temp.next = temp.next.next
        dummy.next = None
        temp = temp.next
        self.count -=1

      else:
        temp = temp.next

    print(f"The linkedd list after removing {target}:",end =" ")
    self.get_ll()



```
