```python
# Power of 3
def check(n):
  # base condition 1 = 3^0
  if n == 1:
    return True
  elif n%3 != 0 or n<=0:
    return False


  else:
    return check(n/3)

check(27)
```

```python
def shift(array):
  if len(array) == 1:
    return array[0]
  else:
    if len(array)!=1:
      array = [array[i] for i in range(len(array)) if i%2==1 ]
    else:
      return array[0]

    if len(array)!=1:
      array = array[::-1]
      array = [array[i] for i in range(len(array)) if i%2==1 ]
      array = array[::-1]
    else:
      return array[0]
    return shift(array)
def get(n):
  array = list(range(1,n+1))
  return shift(array)
  

get(9)
```

The idea is to consider two cases for every character.
(i) Consider current character as part of current subset
(ii) Do not consider current character as part of the current subset.  

Follow the approach to implement the above idea:

The base condition of the recursive approach is when the current index reaches the size of the given string (i.e, index == n). then print the current string say, curr.
Make two recursive calls for the two cases for every character
We consider the character as part of the current subset
We do not consider current character as part of the current subset

```python
def sub(string, index,curr):
  if index == len(string):
    print(curr,end=';')
    return 

  sub(string,index+1,curr+string[index])
  sub(string,index+1,curr)

sub('abc',0,"")
```

```python
def countSubstrs(str, i, j, n):
 
    # base cases
    if (n == 1):
        return 1
    if (n <= 0):
        return 0
 
    res = (countSubstrs(str, i + 1, j, n - 1)
        + countSubstrs(str, i, j - 1, n - 1)
        - countSubstrs(str, i + 1, j - 1, n - 2))    
 
    if (str[i] == str[j]):
        res += 1
 
    return res
 
str = "abcab"
n = len(str)
print(countSubstrs(str, 0, n - 1, n))
```

```python
vowels=['A','E','I','O','U']
def count_cons(s):
  s = s.upper()
  if len(s)==1:
    if ord(s) >= 65 and ord(s) <= 90 :
      if s in vowels:
        return 0
      else:
        return 1
    else:
      return 0
  else:
    if ord(s[0]) >= 65 and ord(s[0]) <= 90 :
      if s[0] in vowels:
        return 0 + count_cons(s[1:])
      else:
        return 1 + count_cons(s[1:])
    else:
      return 0 + count_cons(s[1:])

count_cons('abc de')
```
